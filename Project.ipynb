```python
filter_1=data['CountryName'].str.contains('France')
```

```python
data[filter_1]
```

```python
filter_2=data['IndicatorCode'].str.contains('SP.DYN.IMRT.IN')
```

```python
infant_mortality = data[filter_1 & filter_2]
infant_mortality.head()
```

```python
filter_3=data['IndicatorCode'].str.contains('SH.MED.BEDS.ZS')
```

```python
hospital = data[filter_1 & filter_3]
hospital.head()
```

```python
merged_data=infant_mortality.merge(hospital, on='Year', how='inner')
```

```python
merged_data.tail(15)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(merged_data['Year'].values, merged_data['Value_y'].values)
plt.xlabel('Year')
plt.ylabel(merged_data['IndicatorName_y'].iloc[0])
plt.title('Hospital beds per 1,000 people in France (1974-2011)')
plt.show()
```

```python
plt.plot(merged_data['Year'].values, merged_data['Value_x'].values)
plt.xlabel('Year')
plt.ylabel(merged_data['IndicatorName_x'].iloc[0])
plt.title('Infant mortailty in France (1974-2011)')
plt.show()
```

```python
X=merged_data['Value_y']
Y=merged_data['Value_x']
fig, axis=plt.subplots()
axis.set_title('Infant mortality vs hospital beds per 1,000 people')
axis.set_xlabel(merged_data['IndicatorName_y'].iloc[0])
axis.set_ylabel(merged_data['IndicatorName_x'].iloc[0])
plt.scatter(X,Y)
plt.show()
```

```python
fig, axes=plt.subplots(1,2)
bplot1=axes[0].boxplot(merged_data['Value_y'],notch=True, vert=True, patch_artist=True)
bplot2=axes[1].boxplot(merged_data['Value_x'],vert=True, patch_artist=True)
axes[0].yaxis.grid(True)
axes[0].set_xlabel('Hospital beds')
axes[1].yaxis.grid(True)
axes[1].set_xlabel('Infant mortality')
```
